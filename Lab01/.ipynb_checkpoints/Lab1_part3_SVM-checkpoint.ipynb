## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from mlxtend.plotting import plot_decision_regions

lr = LogisticRegression(penalty='l2', tol=1e-3, C=10, solver='saga', max_iter=1e3) # add some params
svm = SVC(kernel='linear', C=10) # here too

### YOUR CODE HERE
def plot_decisions(clf, X, y):
  clf.fit(X, y)
  plot_decision_regions(X, y, clf = clf, legend = 2)
  print('accuracy: %.2f' % accuracy_score(y, clf.predict(X)))
```

```python
plot_decisions(lr, moons_points, moons_labels)
```

```python
plot_decisions(svm, moons_points, moons_labels)
```

## 1.2 Kernel tirck



Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

* What have happenned with classification quality?
* How did decision border changed for each kernel?
* What `degree` have you chosen and why?

```python
### YOUR CODE HERE
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

# In case we want to find best hyperparams 
def cross_val(model, space):
  search = GridSearchCV(estimator=model, param_grid=space, scoring='accuracy', refit=True, n_jobs=-1)
  CV = search.fit(moons_points, moons_labels)
  print(CV.best_params_)
  return CV 
```

**Poly**

```python
model = SVC(kernel='poly')
space = {'C': np.logspace(-2, 2, 50),
         'degree': np.linspace(1, 10, 10)}
CV = cross_val(model=model, space=space)

plot_decisions(CV.best_estimator_, moons_points, moons_labels)
```

Now lets play with degree since we found the best model:

```python
plot_decisions(SVC(kernel='poly', C=10, degree=2), moons_points, moons_labels)
```

```python
plot_decisions(SVC(kernel='poly', C=10, degree=4), moons_points, moons_labels)
```

Almost the same result we'll get with any even degree (somehow parabolically splitting). It's hard to fit this data with even degrees and much easier with odd. Since CV gives us the best degree is 3, it's worth to check some other odd degrees. 

```python
plot_decisions(SVC(kernel='poly', C=10, degree=5), moons_points, moons_labels)
```

```python
plot_decisions(SVC(kernel='poly', C=10, degree=9), moons_points, moons_labels)
```

It should be noted that here the regularization coefficient penalizes the weights the more the smaller it is and vice versa.

**RBF**

```python
model = SVC(kernel='rbf')
space = {'C': np.logspace(-3, 3, 50)}
CV = cross_val(model=model, space=space)
plot_decisions(CV.best_estimator_, moons_points, moons_labels)
```

Almost perfect.

**Sigmoid**

```python
plot_decisions(SVC(kernel='sigmoid', C=10), moons_points, moons_labels)
```

```python
model = SVC(kernel='sigmoid')
space = {'C': np.logspace(-3, 3, 50)}
CV = cross_val(model=model, space=space)
plot_decisions(CV.best_estimator_, moons_points, moons_labels)
```

For some reason sigmoid can't fit this data and linear border is it's best perfomance (heavily penalized weights).

## 1.3 Simpler solution (of a kind)
What if we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

### YOUR CODE HERE
pipe = Pipeline(steps=[
    ('preprocessor', PolynomialFeatures(degree=2)),
    ('estimator', LogisticRegression(penalty='l2', C=1, tol=1e-3, solver='saga', max_iter=1e3))
    ])

plot_decisions(pipe, moons_points, moons_labels)
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', PolynomialFeatures(degree=3)),
    ('estimator', LogisticRegression(penalty='l2', C=10, tol=1e-3, solver='saga', max_iter=1e3))
    ])

plot_decisions(pipe, moons_points, moons_labels)
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', PolynomialFeatures(degree=4)),
    ('estimator', LogisticRegression(penalty='l2', C=10, tol=1e-3, solver='saga', max_iter=1e3))
    ])

plot_decisions(pipe, moons_points, moons_labels)
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', PolynomialFeatures(degree=5)),
    ('estimator', LogisticRegression(penalty='l2', C=10, tol=1e-3, solver='saga', max_iter=1e3))
    ])

plot_decisions(pipe, moons_points, moons_labels)
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', PolynomialFeatures(degree=9)),
    ('estimator', LogisticRegression(penalty='l2', C=10, tol=1e-3, solver='saga', max_iter=1e3))
    ])

plot_decisions(pipe, moons_points, moons_labels)
```
